```python
#Sterling Vivier PHYS 404
import matplotlib.pylab as p;
from mpl_toolkits.mplot3d import Axes3D
from numpy import *;
import numpy as np;
print("Initializing")
Nmax = 100; Niter = 5000; V = zeros((Nmax, Nmax), float)  

P = zeros((Nmax, Nmax), float) # value to store density
P[50, 50]=1 #density at specific point, 0 everywhere else

for iter in range(Niter):                                  
    maxerr=0.
    for i in range(1, Nmax-1):                                                
        for j in range(1,Nmax-1): 
            # similar code to lecture notes, except now accounting for the right hand side
            # of the equation not being == to 0 :: 4pi*P[i,j]*G. G==1
            du = .25 *(V[i+1,j]+V[i-1,j]+V[i,j+1]+V[i,j-1])-np.pi*P[i,j]*1 -V[i,j]  
            if(maxerr<abs(du)): # how much correction we have in this step
                maxerr=abs(du)  
            V[i,j]= V[i,j] + du 
    if(maxerr<1.e-3):
        break
            
x = range(0, Nmax, 1)
y = range(0, Nmax, 1)                              
X, Y = p.meshgrid(x,y)                 

#setting parameters to plot the 3D function 
def functz(V):                                         # V(x, y) 
    z = V[X,Y]                        
    return z

Z = functz(V)                          
fig = p.figure()                                      # Create figure
ax = Axes3D(fig)                                      # Plot axes
ax.plot_wireframe(X, Y, Z, color = 'r')               # Red wireframe
ax.set_xlabel('X')                                     
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
p.show()      # Show fig

fig = p.figure(figsize=(16,10))
ax = fig.gca()
ax.plot(x, Z)
ax.set_xlabel(r"Position $x$")
ax.set_ylabel(r"Potential")
ax.minorticks_on()


## gravitational potential is known to be negative and to move towards 0 as it moves towards an infinite distance. 
## this is able to be observed in the 3D graph outputted below. 
```
