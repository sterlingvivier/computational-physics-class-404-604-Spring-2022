Burger's equation differs from the advection equation for PDE's due to the advection velocity U not being constant. 
This term produces an output that is non-symmetric over time as seen in the plots below.


```python
# Burger's equation:

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#provided code

# Model Parameters
xmin = -10.0   # left boundary
xmax = +10.0   # right boundary
Nx = 101      # number of grid points (including boundary)
tend = 10.0    # end time
dtout = 1.0   # time interval for outputs

# Set up the grid.
x = np.linspace(xmin, xmax, Nx)
dx = (xmax - xmin) / (Nx - 1)
dt = 0.8 * dx
U = np.zeros(Nx,)

# Give the initial profile.
t = 0.0
U = 0.2 + 0.8 * np.exp(-0.5 * x**2)

# Prepare for 3D outputs.
tp = [t]
Up = np.copy(U)

# Initiate the plot.
#code from lecture notes that set up the plot
fig = plt.figure(figsize=(16,10))
ax = fig.gca()
ax.plot(x, U, label=r"$t = {:.3G}$".format(t))
ax.set_xlabel(r"Position $x$")
ax.set_ylabel(r"$U(t,x)$")
ax.minorticks_on()

# Begin the simulation.
tout = t + dtout
while t < tend:
    # Backup the previous time step.
    Uold = np.copy(U)

    # Find the state at the next time step.
    if (dx/dt) > 0:
        for ix in range(1, Nx - 1):
            #Upwind method
            #similar code to the lecture, however now includes U[ix]*(dx/dt)
            #vel changes with each U[ix]
            U[ix]=Uold[ix]-U[ix]*((dt/dx)*dt) * ((Uold[ix] - Uold[ix-1])/dx)
            
    else:
        for ix in range(1, Nx - 1):
            #Upwind method
            #similar code to the lecture, however now includes U[ix]*(dx/dt)
            #vel changes with each U[ix]
            U[ix]=Uold[ix] -U[ix]*((dt/dx)*dt) *((Uold[ix +1] - Uold[ix])/dx)          
    t += dt

    # Save the data after every dtout.
    if t >= tout:
        plt.plot(x, U, label=r"$t = {:.3G}$".format(t))
        tp.append(t)
        Up = np.vstack((Up, U))
        print("t = ", t)
        tout += dtout

ax.legend()

#Create 3D-view of the solution.
t, x = np.meshgrid(tp, x)
fig3D = plt.figure(figsize=(16,10))
ax3D = Axes3D(fig3D)
ax3D.plot_wireframe(t, x, Up.transpose(), color="red")
ax3D.set_xlabel("Time $t$")
ax3D.set_ylabel("Position $x$")
ax3D.set_zlabel(r"$U(t,x)$")

print("Done.")
plt.show()
```
