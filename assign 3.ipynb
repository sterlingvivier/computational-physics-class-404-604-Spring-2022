```python
# Sterling Vivier Phys 404 Assignment 3

import numpy as np #importimg necessary libraries
import matplotlib.pyplot as plt

def derifw(y,x,dx): #forward method
    return (y(x+dx)-y(x))/dx

x=np.pi/4 # pi/4
dx=0.1
dxarr=[] #creates array
yarr=[]

for dx in 10**np.arange(-16., -1, 0.1): # range
    y=derifw(np.cos,x,dx)
    dxarr.append(dx)
    yarr.append(y)
    
plt.plot(dxarr, abs(yarr--np.sin(x)), 'b-', lw=2)    #as dx gets smaller, the error gets smaller
plt.xscale('log')     # error is 10^-8
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')

```

```python
def dericd(y,x,dx): #central difference 
    return (y(x+dx/2)-y(x-dx/2))/dx

x=np.pi/4 #pi/4
dx=0.1
dxarr2=[] #creates array
yarr2=[]


for dx in 10**np.arange(-16., -1, 0.1): #range
    y=dericd(np.cos,x,dx)
    dxarr2.append(dx)
    yarr2.append(y)
    
plt.plot(dxarr, abs(yarr2--np.sin(x)), 'b-', lw=2)    #as dx gets smaller, the error gets smaller
plt.xscale('log')      #compared to first method, this one is more accurate
plt.yscale('log')      # steeper slope indicates more accurate scheme
plt.xlabel('dx')
plt.ylabel('Error') # error is 10^-5
```
