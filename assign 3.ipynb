```python
# Sterling Vivier PHYS 404 Assignment 3
import numpy as np
import matplotlib.pyplot as plt

def derifw(y,x,dx): #forward method 
    return (y(x+dx)-y(x))/dx

x= np.pi/4  #x value

dxarr=[]  #arrays to hold variables
yarr=[]

for dx in 10**np.arange(-16, -1, 0.1): #x range 
    y=derifw(np.cos,x,dx)
    dxarr.append(dx) #appends values to arrays
    yarr.append(y)
    
plt.plot(dxarr, abs(yarr--np.sin(np.pi/4))) #plot error
plt.xlabel('dx')
plt.xscale('log')
plt.yscale('log')
plt.ylabel('Error')
plt.show()

#error is 10^-8.
```

```python
def dericd(y,x,dx): #central difference 
    return (y(x+dx/2)-y(x-dx/2))/dx

x= np.pi/4 #central difference 

dxarr2=[]  #arrays for variables 
yarr2=[]

for dx in 10**np.arange(-16, -1, 0.1): #range 
    y=dericd(np.cos,x,dx)
    dxarr2.append(dx) #appends values to arrays
    yarr2.append(y)
    
plt.plot(dxarr2, abs(yarr2--np.sin(x))) # plot error
plt.xscale('log') 
plt.yscale('log')
plt.xlabel('dx')
plt.ylabel('Error')
plt.show()

#slope is steeper, thus scheme is more accurate. 
#error is 10^-5.
#with this method, dx is able to go smaller
```
