```python
#Sterling Vivier PHYS 404
#Homework2

import matplotlib.pyplot as plt #import matplot library

fiboCount={} #varaibles to store the values of f(n) & f(n-1)
fiboCount2={}

def fibo(n): #function 1 that finds f(n)
    if n in fiboCount: #memorization method checks to see if the
        return fiboCount[n] #value has already been calculated
    if n==1:
        value=1
    if n==2:
        value=1
    if n>2:
        value=fibo(n-1) + fibo(n-2)
    
    fiboCount[n]=value
    return value #function will return value calculated

def fibo2(n): #function 2 that finds f(n-1)
    if n in fiboCount2: #calculates in a similar manner, just starting at 0 && not 1
        return fiboCount2[n]  
    if n==1:
        value2=0
    if n==2:
        value2=1
    if n>2:
        value2=fibo2(n-1) + fibo2(n-2)
        
    fiboCount2[n]=value2
    return value2 #function will return value calculated

y=[] #arrays to store the calculated values
x=[]

for n in range (2, 55): #loop calls the function and performs the necessary calculation
    y.append(fibo(n)/fibo2(n))
    x.append(n)

plt.plot(x,y) #plots calculated f(n)/f(n-1)
plt.xlabel("n")
plt.ylabel("F(n)/F(n-1)")
plt.show()



    
```

```python
#TABLE METHOD

#The graph generated from using this method has a complexity of O(logn).
#As n grows, the graph grows logarithmically proportional

import matplotlib.pyplot as plt #importing necessary libraries
import time
import math
import numpy as np
import os
    
def fibT(n): #calculates the fibonacci sequence using the table method
    table=[0]*(n+1)
    table[n] = 0
    table[1] = 1
    
    for i in range (2, n+1):
        table[i]=  table[i-1] + table[i-2]
        
    return table[n]


file1= open('table.txt', 'w') #creates/opens a file to output n and time
x=np.arange(5, 1000, 10) #sets x/n. starts at 5 and goes to 1000 in 10increments
time2=np.zeros(x.size)
for i in range(x.size): #for loop that calculates the time && outputs results to a file
    start_time = time.time()
    temp=fibT(x[i])
    time2[i] = time.time()-start_time
    string=str(time2[i])
    file1.write(str(x[i]))
    file1.write(" ")
    file1.write(string)
    file1.write("\n")

file2 = open('table.txt', 'r') #variable to read from the created file

x1=[] #array variables to hold file data
x2=[]

for line in file2:
        line = line.strip() #removes characters
        data = line.split() #splits line into a list
        x1=np.append(x1, float(data[0])) #breaks up elements of the list to
        x2=np.append(x2, float(data[1])) #store in the array

plt.plot(np.log10(x1), np.log10(x2)) #log graph of the results n vs time
```

```python
#RECURSION METHOD

#Using recursion method, the time needed to calculate this sequence
#quickly becomes very long even while n is still relatively small
#the slowest method of the three

import matplotlib.pyplot as plt  #importing necessary libraries
import time
import math
import numpy as np
import os

def fibR(n): #calculates the fibonacci sequence using the recursion method
    if n==0:
        return 0
    if n==1:
        return 1
    if n==2:
        return 1
    if n>2:
        return fibR(n-1) + fibR(n-2)

file5=open('recursion.txt', 'w') #creates/opens a file to output the results of the recursion method

x=np.arange(1, 30, 1) #arranges x/n. starts at 1 and goes to 30 in 1 increments
                      #calculating the function after this x runs very slowly
time2=np.zeros(x.size)

for i in range(x.size): #for loop that calculates the time && outputs results to a file
    start_time = time.time()
    temp=fibR(x[i])
    time2[i] = time.time()-start_time
    string=str(time2[i])
    file5.write(str(x[i]))
    file5.write(" ")
    file5.write(string)
    file5.write("\n")
    
file6 = open('recursion.txt', 'r') #variable to read from the created file

x1=[] #array variables to hold file data
x2=[]

for line in file6:
        line = line.strip() #removes characters
        data = line.split() #splits line into a list
        x1=np.append(x1, float(data[0])) #breaks up elements of the list to
        x2=np.append(x2, float(data[1])) #store in the array

plt.plot(np.log10(x1), np.log10(x2)) #log graph of the results n vs time
```
